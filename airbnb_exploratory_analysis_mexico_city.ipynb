#### **Identifying missing data**

Primero, realizamos un análisis para identificar las variables que contienen valores faltantes. Esta etapa es importante para comprender la magnitud del problema y su impacto en nuestro análisis. Después, evaluamos diferentes estrategias para manejar los datos ausentes, que pueden incluir desde la eliminación de filas o columnas con valores faltantes hasta técnicas de imputación que nos permiten estimar valores basados en información disponible.

Para identificar los valores faltantes en cada variable, creamos un DataFrame con valores booleanos que indiquen la presencia de valores faltantes en cada celda de nuestro conjunto de datos.

```python
missing_data = df.isna()
display(missing_data)
```

Los valores boolenaos indican la presencia (True) o ausencia (False) de valores faltantes en cada celda de nuestro conjunto de datos.

Excluimos las variables que no tienen ningún valor True en el DataFrame `missing_data`, es decir, nuestro DataFrame `missing_data` contendrá solo las columnas que tienen al menos un valor faltante.



```python
missing_data = missing_data.loc[:, missing_data.any(axis= 0)]
display(missing_data)
```

#### **Counting missing data**

Usando un bucle `for` en Python, podemos calcular rápidamente la cantidad de valores faltantes en cada variable de `missing_data` y el porcentaje que representan. 

```python
for column in missing_data.columns:
    print(" ")
    print(f"Variable: {column}")
    print(f"Missing Values: {missing_data[column].sum()} de {len(missing_data)}")
    print(f"% NaNs: {(missing_data[column].sum()/len(missing_data)*100).round(2)}")
```

##### **Handle numeric variables**

A continuación se abordará el tratamiento de valores faltantes en las variables numéricas de nuestro conjunto de datos. Se analizarán diferentes formas o técnicas para tratar los valores faltantes, como la imputación con la media o la mediana, la imputación basada en información adicional, entre otras, con el objetivo de garantizar que la ausencia de datos se refleje de manera clara y precisa en el análisis final.

Primero, separamos nuestras variables numéricas y categóricas que contienen valores NaN para tratarlos por separado.

```python
# crear DataFrame de variables númericas con valores NaN

missing_data_numeric = df.select_dtypes(include = 'number').isna()
missing_data_numeric = missing_data_numeric.loc[:, missing_data_numeric.any(axis = 0)]
missing_data_numeric
```

Calculamos la cantidad de valores faltantes en cada variable numérica de `missing_data_numeric` y el porcentaje que representan. 

```python
for column in missing_data_numeric.columns:
    print(" ")
    print(f"Variable: {column}")
    print(f"Missing Values: {missing_data_numeric[column].sum()} de {len(missing_data_numeric)}")
    print(f"% NaNs: {(missing_data_numeric[column].sum()/len(missing_data_numeric)*100).round(2)}")
```

Para las variables `bedrooms`, `bathrooms` y `beds` optamos por imputar los valores faltantes utilizando la mediana. Esta elección se basa en que la mediana ofrece una solución robusta al ser menos sensible a valores extremos (que son comunes en conjuntos de datos de este tipo). Cuando hablamos de la cantidad de habitaciones (bedrooms), baños (bathrooms) o camas (beds) en alojamientos, es común que estos valores se distribuyan de manera no uniforme, es decir, que haya más alojamientos con cierta cantidad de habitaciones o baños que con otras cantidades. En este sentido, la mediana, al ser una medida que se encuentra en el punto medio de un conjunto de datos ordenados, refleja mejor cuál es la cantidad más típica de habitaciones, baños o camas que se encuentra en nuestra muestra de alojamientos. También de esta manera se evita la introducción de fracciones en variables que naturalmente se expresan en números enteros

Por lo tanto, al utilizar la mediana para imputar los valores faltantes en  `bedrooms`, `bathrooms` y `beds`, aseguramos una aproximación más robusta y representativa de la distribución típica de estas características en nuestra muestra de alojamientos.

```python
# Mediana de 'bedrooms', 'bathroom's y 'beds'

df[['bedrooms','bathrooms','beds']].median()
```

```python
# Imputar valores NaN en 'bedrooms' 

df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
```

```python
# Imputar valores NaN en 'bathrooms' 

df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())
```

```python
# Imputar valores NaN en 'beds' 

df['beds'] = df['beds'].fillna(df['beds'].median())
```

##### **Handle categorical variables**

Para tratar los valores faltantes en nuestras variables categóricas podemos considerar opciones como reemplazarlos con el valor que aparece con mayor frecuencia o la creación de categorías especiales para representar los valores NaN, entre otras.

```python
# crear DataFrame de variables categóricas con valores NaN

missing_data_categoric = df.select_dtypes(include = 'object').isna()
missing_data_categoric = missing_data_categoric.loc[:, missing_data_categoric.any(axis = 0)]
missing_data_categoric
```

Calculamos la cantidad de valores faltantes en cada variable categorica que contiene al menos un valor NaN y el porcentaje que representan. 

```python
for column in missing_data_categoric.columns:
    print(" ")
    print(f"Variable: {column}")
    print(f"Missing Values: {missing_data_categoric[column].sum()} de {len(missing_data_categoric)}")
    print(f"% NaNs: {(missing_data_categoric[column].sum()/len(missing_data_categoric)*100).round(2)}")
```

Podemos observar la presencia de valores NaN en varias variables categóricas. Tras evaluar la proporción de los valores faltantes en las variables `host_response_time`, `host_has_profile_pic`, `host_is_superhost` y `host_identity_verified`, hemos decidido imputarlos utilizando la moda en lugar de crear categorías especiales. Esta decisión se basa en el hecho de que los valores faltantes no representan una proporción significativa de los datos, y la imputación por la moda nos proporcionará estimaciones mas precisas basadas en los valores más frecuentes observados en nuestro conjunto de datos. La presencia de una categoría especial podría introducir ambigüedad en la interpretación de los datos y dificultar la comparación entre los grupos de las variables.

```python
# Imputar valores NaN en 'host_response_time'

df['host_response_time'] = df['host_response_time'].replace(np.NaN, df['host_response_time'].mode()[0])
```

```python
# Imputar valores NaN en 'host_has_profile_pic'

df['host_has_profile_pic'] = df['host_has_profile_pic'].replace(np.NaN, df['host_has_profile_pic'].mode()[0])
```

#### **Skewness and Kurtosis**

La asimetría (skewness) y la curtosis (kurtosis) son medidas estadísticas que proporcionan información sobre la forma y las características de una distribución de datos. Proporcionan medidas cuantitativas que ayudan a comprender la forma, simetría y características específicas de una distribución. Y son útiles para identificar la presencia de sesgo, colas pesadas, o características que difieren de una distribución normal.

La skewness mide la asimetría o la falta de simetría en una distribución de datos.

Interpretación:
* Skewness positiva: La cola derecha de la distribución es más larga o pesada que la cola izquierda. La mayoría de los datos se concentran en el lado izquierdo.
* Skewness negativa: La cola izquierda de la distribución es más larga o pesada que la cola derecha. La mayoría de los datos se concentran en el lado derecho.
* Skewness cercana a cero: La distribución es aproximadamente simétrica.

Una cola larga puede referirse tanto a una mayor probabilidad de observar valores extremos como a la extensión física de la cola en sí. Una cola larga indica que la cola se extiende más lejos de la media en comparación con una distribución normal.

La kurtosis mide la "tailedness" o "pesadez de la cola" de una distribución de datos e indica el grado de concentración de la distribución alrededor de su media. La pesadez de la cola se refiere a una mayor o menor probabilidad de observar valores extremos en las colas de la distribución en comparación con lo que se esperaría en una distribución normal. En una cola pesada, la concentración de datos en la región central es alta, pero hay una probabilidad relativamente alta de observar valores extremos lejos de la media.

Interpretación:
* Kurtosis positiva: Colas más pesadas y una distribución más puntiaguda (leptocúrtica) que la distribución normal.
* Kurtosis negativa: Colas más ligeras y una distribución más achatada (platicúrtica) que la distribución normal.
* Kurtosis igual a cero: La distribución tiene la misma pesadez de cola que la distribución normal (mesocúrtica).

Mientras que la curtosis se centra en las colas y la concentración de datos, la skewness se enfoca en la simetría o asimetría direccional de la distribución. 

En nuestro análisis univariante de la variable 'price', calcular las medidas de asimetría y curtosis antes de visualizar la distribución puede proporcionar un contexto inicial sobre la forma general de la distribución.

Para calcular la asimetría y la curtosis utilizamos las funciones estadísticas en Python `stats.skew()` y `stats.kurtosis()`

```python
# Calcular la asimetría y la curtosis

skew_price = stats.skew(df['price']).round(2)
kurtosis_price = stats.kurtosis(df['price']).round(2)

print(f"Skewness= {skew_price}, Kurtosis= {kurtosis_price}")
```

* **Asimetría (skewness): 8.4**  
Un valor de asimetría positivo indica sesgo a la derecha. En este caso, la cola derecha de la distribución es más larga o pesada que la cola izquierda. En términos prácticos, esto sugiere que hay valores más altos o extremos en el lado derecho de la distribución, lo que podría indicar la presencia de algunos alojamientos con precios significativamente más altos. La asimetría positiva sugiere que la distribución de los precios está sesgada a la derecha, con la presencia de algunos valores extremadamente altos.

* **Curtosis (kurtosis): 137.07**  
Un valor de curtosis positivo indica que la distribución tiene colas más pesadas y es más puntiaguda que una distribución normal (leptocúrtica), lo que podría indicar la presencia de valores extremos y una mayor concentración de datos alrededor de la media. Esto significa que hay una mayor concentración de datos en la región central de la distribución y una mayor probabilidad de observar valores extremos en las colas, especialmente en el lado derecho debido a la asimetría positiva.

Estas interpretaciones nos proporcionan una visión general de la forma de la distribución de la variable `price`, proporcionando información sobre la concentración de datos alrededor de la media, la presencia de valores extremos y la simetría de la distribución. Estos límites indican la región dentro de la cual se espera que se encuentren la mayoría de los datos, y cualquier punto fuera de estos límites se consideraría un posible valor atípico o extremo. #### **Data Grouping**

El Data Grouping o Agrupamiento de Datos es una técnica utilizada en análisis de datos para organizar y estructurar la información en conjuntos o grupos basados en ciertas características o criterios comunes, para luego realizar cálculos o análisis dentro de esos grupos. Esto se hace con el objetivo de facilitar el análisis y la interpretación de los datos al identificar patrones, tendencias o relaciones dentro de cada grupo.

En nuestro proyecto, el agrupamiento de datos puede aplicarse a variables categóricas o numéricas para generar insights significativos. Por ejemplo, en el análisis de precios de alojamientos de Airbnb en la Ciudad de México, podemos agrupar los alojamientos por tipo de habitación (como casa completa, habitación privada, habitación compartida) o por ubicación (por alcaldía o zona de la ciudad), lo que permitiría realizar comparaciones y extraer conclusiones más específicas sobre cada grupo.

En Python, la biblioteca Pandas proporciona el método `groupby()` que es muy útil para realizar operaciones de agrupación en un DataFrame. El método `groupby` agrupa los datos por diferentes categorías. Los datos se agrupan en función de una o varias variables y el análisis se realiza en los grupos individuales.

Podemos conocer la distribución de la variable `price` en función de las categorías de una variable categórica utilizando `groupby` y `mean()` para calcular la media de `price` para cada grupo individual.

A continuación, agruparemos los alojamientos por alcaldía (`neighbourhood`) y calcularemos la media de los precios en cada grupo. Esta acción nos permitirá identificar y comparar los precios medios de alojamiento en las diferentes alcaldías de la Ciudad de México. Posteriormente, ordenaremos estos resultados de manera descendente para destacar las alcaldías que presentan los precios promedio más altos en el mercado de Airbnb de la Ciudad de México.

```python
df.groupby('neighbourhood', as_index= False)['price'].mean().round(2) \
    .sort_values(by= 'price', ascending= False).reset_index(drop= True)
```

Los resultados muestran el precio medio de alojamientos en cada alcaldía de la Ciudad de México. Estos valores representan un promedio de los precios de alojamiento por noche en cada área específica. Al analizar estos datos, podemos observar diferencias significativas en los precios entre las diferentes alcaldías.

* Miguel Hidalgo, con un precio medio de $1708.17, muestra una tendencia hacia alojamientos más costosos en comparación con otras áreas. Cuauhtémoc, con un precio medio de $1594.87, también presenta precios relativamente altos en alojamientos, aunque ligeramente inferiores a los de Miguel Hidalgo. Las posiciones de la alcaldía Cuauhtémoc y Miguel Hidalgo se podrían explicar por abarcar la zona centro de la CDMX. Cuajimalpa de Morelos, con un precio medio de $1324.94, se encuentra en un rango intermedio en términos de precios de alojamientos.

* Iztapalapa, con un precio medio de $616.76, representa una de las opciones más económicas en cuanto a alojamientos en la Ciudad de México. Tláhuac, con un precio medio de $396.79, muestra precios aún más bajos en comparación con las demás alcaldías, siendo una opción a considerar para aquellos con un presupuesto limitado.

Estos resultados son útiles para entender la distribución de precios de alojamientos de Airbnb en la Ciudad de México y pueden ser de interés para viajeros que buscan opciones de alojamiento dentro de un determinado presupuesto o área geográfica específica.

#### **Correlation Analysis**



El análisis de correlación es una técnica estadística que se utiliza para evaluar la relación entre dos o más variables numéricas. Su objetivo principal es determinar si existe una asociación entre las variables y en qué medida se relacionan, en este caso, entre nuestras variables continuas como la cantidad de huéspedes que permite el anfitrion, el número de habitaciones, camas, baños, etc., y el precio de alquiler de los alojamientos.

El análisis de correlación proporciona un valor numérico llamado coeficiente de correlación, denotado como r, el coeficiente de correlación es la medida que cuantifica la fuerza y la dirección de una relación lineal entre dos variables. En otras palabras, cuando observamos dos variables, si una variable cambia, ¿cómo afecta esto al cambio en la otra variable? Cuando dos variables están correlacionadas positivamente, significa que tienden a cambiar en la misma dirección, es decir, cuando una variable aumenta, la otra también tiende a aumentar. Por otro lado, si las variables están correlacionadas negativamente, significa que tienden a cambiar en direcciones opuestas; cuando una variable aumenta, la otra tiende a disminuir.

El coeficiente de correlación puede tomar un rango de valores entre -1 y 1 inclusive, donde:

El valor absoluto del coeficiente de correlación de Pearson |r| indica la fuerza de la asociación.

* Si |r| está cerca de 1, la correlación es fuerte entre las variables.
* Si |r| está cerca de 0, la correlación es débil o inexistente entre las variables.

El signo de r indica la dirección de la asociación.

* Si r > 0 indica una correlación positiva
* Si r < 0 indica una correlación inversa o negativa

Entonces:

* Si **r = 1:** Correlación lineal positiva perfecta.
* Si **r = 0:** Sin correlación lineal, lo más probable es que las dos variables no se afecten entre sí.
* Si **r = -1:** Correlación lineal negativa perfecta.


Al realizar el análisis de correlación, se buscarán relaciones significativas entre las variables numéricas y el precio de los alojamientos. Un coeficiente de correlación cercano a 1 o -1 indicaría una fuerte asociación lineal entre las variables, lo que podría indicar que esa variable tiene un impacto considerable en el precio de los alojamientos. Por otro lado, un coeficiente de correlación cercano a 0 sugeriría una relación débil o nula entre las variables.

Este análisis de correlación roporcionará insights sobre qué variables numericas tienen una relación más estrecha con el precio de los alojamientos en la Ciudad de México, lo que ayudará a comprender mejor los factores que influyen en los precios de los alojamientos en este mercado.

Para calcular el coeficiente de correlación entre la variable `price` y las variables numéricas de nuestro conjunto de datos, se empleará el método `corr()`. La correlación de Pearson es el método predeterminado utilizado por esta función.

```python
# Crear DataFrame unicamente con las variables numricas de nuestro conjunto de datos

numerical_columns = df.select_dtypes(include= 'number')
numerical_columns = numerical_columns.drop(columns= ['latitude', 'longitude'], errors='ignore')
numerical_columns.head(5)
```
